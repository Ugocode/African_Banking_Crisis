## Using SMOTE

Ill up-sample the crisis using the SMOTE algorithm(Synthetic Minority Oversampling Technique). At a high level, SMOTE: Works by creating synthetic samples from the minor class (crisis) instead of creating copies. Randomly choosing one of the k-nearest-neighbors and using it to create a similar, but randomly tweaked, new observations.

```python
# First prepear the data 
data['banking_crisis'] = data['banking_crisis'].map({'no_crisis': 0, 'crisis': 1})
```

```python
data.banking_crisis.head(4)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X = data.drop(['banking_crisis'], axis = 1)
```

```python
X.head(3)
```

```python
y = data.banking_crisis
```

```python
y.head(4)
```

I  am dropping the cc3 which is simply a  3 letter country code column from the dataset and also the year which might cause problem for our machine learning algorithm  

```python
####  Droping the irrelevant features 
```

```python
to_drop = ['case', 'cc3', 'year', 'country']
```

```python
X.drop(to_drop, axis = 1, inplace = True)
X.head(3)
```

## Splitting and Training Dataset

```python
# Importing the sklearn train test split module
from sklearn.model_selection import train_test_split
```

```python
# splitting my dataset to get the Training and Testing Data 

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.20, random_state = 123, stratify = y)
```

```python
# Setting up the upsampling method with SMOTE
sm = SMOTE(random_state = 12, ratio = 1.0)  
```

```python
# Over sampling the train data only 
X_train_sm, y_train_sm = sm.fit_sample(X_train, y_train)
```

```python
print("Original shape:", X_train.shape, y_train.shape, '\n')
```

```python
print ("SMOTE sample shape:", X_train_sm.shape, y_train_sm.shape, '\n')
```

As you can see our data was really up sampled with the **SMOTE** method 

 I will be using both **Logistic regression** and **Random Forest Classifier** in making my model 

## Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression(max_iter=150)
```

```python
lr_model.fit(X_train_sm, y_train_sm)
```

```python
crisis_predict = lr_model.predict(X_test)
crisis_predict
```

```python
print('Model Accuracy is ', lr_model.score(X_test, y_test))
```
